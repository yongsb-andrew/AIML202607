## 2. Boxplots

Boxplots, also called box-and-whisker plots, help us visualise the distribution of a variable. They can show:

- the middle value, also called the median
- the lower and upper quartiles
- the interquartile range, or IQR
- possible outliers

A possible outlier is usually a value that is:

- greater than `75th percentile + 1.5 × IQR`, or
- smaller than `25th percentile - 1.5 × IQR`

You can refer to the figure below to understand how to read a boxplot. Optional video: [Boxplot explanation](https://www.youtube.com/watch?v=o7qWblT5NZI)

<img src="./resources/Box_Plot.PNG">

Let us plot a boxplot for the `sepal_length` variable first. The code is provided as a worked example.

> **Note:** `showfliers=True` means possible outlier points will be shown on the plot.

```python
plt.boxplot(df["sepal_length"], showfliers=True, label=["sepal length"])
plt.title("Boxplot of Sepal Length")
plt.ylabel("Length")
plt.show()
```

## 3. Histograms

Histograms also help us visualise the distribution of data. A histogram groups values into intervals called **bins**.

Histograms can help us answer questions such as:

- Which values appear most often?
- Is the data balanced or skewed?
- Are there unusual gaps or peaks?

Try the worked example below for `sepal_length`.

```python
plt.hist(df["sepal_length"], bins=8, range=(4, 8))
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

From the histogram above, the distribution is relatively balanced, with most values falling between about 4.5 and 7.0.

Before choosing the range and number of bins for a histogram, it helps to print the minimum and maximum values. The example below does this for `sepal_length`.

```python
print("min =", df["sepal_length"].min())
print("max =", df["sepal_length"].max())
```

Now print the minimum and maximum values for the remaining variables. This will help you choose suitable histogram ranges.

```python
print("sepal_width min =", df["sepal_width"].min())
print("sepal_width max =", df["sepal_width"].max())

print("petal_length min =", df["petal_length"].min())
print("petal_length max =", df["petal_length"].max())

print("petal_width min =", df["petal_width"].min())
print("petal_width max =", df["petal_width"].max())
```

Now plot histograms for the remaining variables.

Try different bin numbers and ranges to get intervals that are easy to read, such as 1.0, 0.5, or 0.2.

```python
# Histogram for sepal_width
plt.hist(df["sepal_width"], bins=8, range=(2, 4.5))
plt.title("Histogram of Sepal Width")
plt.xlabel("Sepal Width")
plt.ylabel("Frequency")
plt.show()

# Histogram for petal_length
plt.hist(df["petal_length"], bins=8, range=(1, 7))
plt.title("Histogram of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()

# Histogram for petal_width
plt.hist(df["petal_width"], bins=8, range=(0, 3))
plt.title("Histogram of Petal Width")
plt.xlabel("Petal Width")
plt.ylabel("Frequency")
plt.show()
```

From the histograms, you should be able to observe that the distributions for both `petal_length` and `petal_width` are skewed. This supports what we observed earlier from the boxplots.

## 4. Scatter plots

Boxplots and histograms help us study **one variable at a time**.

Scatter plots help us study the relationship between **two variables**. They can help us identify:

- possible outliers involving two variables
- clusters of points
- relationships between variables

Try the code below. It plots `petal_length` against `petal_width`.

```python
plt.scatter(df["petal_length"], df["petal_width"])
plt.xlabel("petal length")
plt.ylabel("petal width")
plt.title("Petal Length vs Petal Width")
plt.show()
```

Do you notice any outlier in the scatter plot? Where are the unusual-looking points? Explain briefly.

```python
# Reference answer:
# The points on the left look separated from the rest, but they form a cluster rather than individual outliers.
# There is a positive relationship between petal length and petal width.
```

At first, it may seem that there are outliers on the left side of the plot because these points are far away from the majority of the points. However, they are probably **not outliers** because they cluster together.

A single unusual point may be an outlier. A group of similar points may be a cluster.

You may also notice a direct relationship between `petal_length` and `petal_width`: when one increases, the other tends to increase as well.

## 4.1 Bonus: Scatter plots by flower class

We can also use scatter plots to understand differences between classes in the dataset.

In the Iris dataset, the `class` column tells us the flower class. Later in machine learning, this will be the label we try to predict.

The code below starts by splitting the dataframe into the 3 flower classes. Complete the missing parts, then plot each class using a different colour.

```python
# Split the dataframe into the 3 classes
df_Setosa = df[df["class"] == "Iris-setosa"]
df_Versicolor = df[df["class"] == "Iris-versicolor"]
df_Virginica = df[df["class"] == "Iris-virginica"]

# Plot the different classes using different colours
plt.scatter(df_Setosa["petal_length"], df_Setosa["petal_width"], c="r", label="Setosa")
plt.scatter(df_Versicolor["petal_length"], df_Versicolor["petal_width"], c="g", label="Versicolor")
plt.scatter(df_Virginica["petal_length"], df_Virginica["petal_width"], c="b", label="Virginica")

plt.legend()
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Length vs Petal Width by Class")
plt.show()
```

This plot shows that `petal_width` and `petal_length` may be important variables if we want to predict flower classes. It also explains why the points at the left corner are not outliers: they belong to the Setosa class.

Now try the same idea using `sepal_length` and `sepal_width`. Is it easy to separate the classes?

```python
# Split the dataframe into the 3 classes
df_Setosa = df[df["class"] == "Iris-setosa"]
df_Versicolor = df[df["class"] == "Iris-versicolor"]
df_Virginica = df[df["class"] == "Iris-virginica"]

# Plot the different classes using different colours
plt.scatter(df_Setosa["sepal_length"], df_Setosa["sepal_width"], c="r", label="Setosa")
plt.scatter(df_Versicolor["sepal_length"], df_Versicolor["sepal_width"], c="g", label="Versicolor")
plt.scatter(df_Virginica["sepal_length"], df_Virginica["sepal_width"], c="b", label="Virginica")

plt.legend()
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Sepal Width by Class")
plt.show()
```
